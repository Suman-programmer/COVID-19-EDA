## Importing Library

```python
import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter,DayLocator
import matplotlib.dates as mdates
import matplotlib.style
import seaborn as sns
sns.set(style="white")
matplotlib.style.use('ggplot')
import matplotlib.colors as mcolors
from datetime import datetime
sns.set(style="white")
import random
```

## Loading Dataset

```python
df = pd.read_csv("train.csv")
df.head()
```

### Data Info 

```python
df.info()
```

### Countries and its total no. of Cases 

```python
unique_countries = df['Country_Region'][df.ConfirmedCases >0 ].unique()
country_confirmed_cases = []
for i in unique_countries:
    country_confirmed_cases.append(df[df.ConfirmedCases > 0 ][df['Country_Region'] == i].ConfirmedCases.sum())
```

```python
for i in range(len(unique_countries)):
    print (f'{unique_countries[i]}: {country_confirmed_cases[i]} cases')
```

```python
c = random.choices(list(mcolors.CSS4_COLORS.values()),k = len(unique_countries))
plt.figure(figsize=(15,15))
plt.pie(country_confirmed_cases, colors = c)
plt.legend(unique_countries, loc='best')
plt.show()
```

## Countries with its total number of Death Cases

```python
country_death_cases = []
for i in unique_countries:
    country_death_cases.append(df[df.Fatalities > 0 ][df['Country_Region'] == i].Fatalities.sum())
```

```python
for i in range(len(unique_countries)):
    print (f'{unique_countries[i]}: {country_death_cases[i]} cases')
```

```python
c = random.choices(list(mcolors.CSS4_COLORS.values()),k = len(unique_countries))
plt.figure(figsize=(15,15))
plt.pie(country_death_cases, colors = c)
plt.legend(unique_countries, loc='best')
plt.show()
```

## Confirm Cases over Days

## Death Cases over Days

```python
sns.relplot(x="Date",y="Fatalities",kind="line",data=df)
plt.show()
```

## Confirmed Cases Vs Fatality Cases

```python
plt.figure(figsize=(15, 10))
sns.relplot(x='ConfirmedCases',y='Fatalities', data=df,kind="scatter",hue="ConfirmedCases")
plt.show()
```

## Top 10 countries with Highest Confirmed Cases

```python
countryDict = {'Country':unique_countries,'no_of_cases':country_confirmed_cases}
countryDf = pd.DataFrame(countryDict)
sorted_countryDf = countryDf.sort_values('no_of_cases', axis=0, ascending=False, inplace=False)
highest_cases = sorted_countryDf.iloc[0:11, :].reset_index(drop=True)
sns.barplot(x="Country", y="no_of_cases",data=highest_cases)
plt.show()
```

## Top 10 countries with Highest Death Cases

```python
country_death_Dict = {'Country':unique_countries,'no_of_cases':country_death_cases}
countryDf_death = pd.DataFrame(country_death_Dict)
sorted_countryDf_death = countryDf_death.sort_values('no_of_cases', axis=0, ascending=False, inplace=False)
highest_deaths = sorted_countryDf_death.iloc[0:11, :].reset_index(drop=True)
sns.barplot(x="Country", y="no_of_cases",data=highest_deaths)
plt.show()
```

## Active Cases Over Days

```python
df = df.drop(df[(df['ConfirmedCases'] == 0.0) & (df['Fatalities'] == 0.0)].index)
df['Active_cases'] = df['ConfirmedCases'] - df['Fatalities']
df.reset_index(drop=True).head()
```

```python
sns.relplot(x="Date",y="Active_cases",kind="line",data=df)
plt.show()
```
